```python
def newton(x,iterations,theta,y):
  costs=[]
  print("x= ",x.shape, " theta= ",theta.shape)
  for i in range(iterations):
    sig = 1/(1+np.exp(-x.dot(theta)))
    #print("sig ",sig.shape)
    cost = -(1/(x.shape[0]))*sum(y*np.log(sig) + (1-y)*np.log(1-sig))
    #print(cost)
    costs.append(cost)
    grad = (1/(np.size(x,0)))*(x.T.dot((sig-y)))
    # hessian matrix
    H = (1/(np.size(x,0)))*(x.T.dot(np.diag(sig.reshape(np.size(x,0),))).dot(np.diag((1-sig).reshape(np.size(x,0),))).dot(x))
    theta = theta - np.linalg.pinv(H).dot(grad)
    print(cost)
  return costs,theta

def plotgraph(costs):
  plt.plot(costs)
  plt.xlabel("Iterations")
  plt.ylabel('cost')
  plt.title("Cost vs Iterations")
  plt.show()

def compute_predictions(X, theta):
  exp_component = [np.exp(-((theta[1] * xi) + theta[0])) for xi in X]    
  ypred =  [e/(1 + e) for e in exp_component] 
  return ypred

def accuracy(Ytrue,Ypred):
  acc=0;
  for i in range(len(Ytrue)):
    if(Ytrue[i] == Ypred[i]):
      acc+=1;
  return acc/len(Ytrue)
```

# **LOGISTIC REGRESSION**

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,accuracy_score,confusion_matrix

def sigmoid(z):
    return  1 / (1 + np.exp(-z))

def cost(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()

def gradient(X, h, y):
    return np.dot(X.T, (h - y)) / y.shape[0]

def logistic_regression(X, y, theta, alpha, iters):
    cost_array = np.zeros(iters)
    for i in range(iters):
        h = sigmoid(np.dot(X, theta))
        cost_num = cost(h, y)
        cost_array[i] = cost_num
        gradient_val = gradient(X, h, y)
        theta = theta - (gradient_val * alpha)
        print(cost_num)
    return theta, cost_array

def plotChart(iterations, cost_num):
    plt.plot(np.arange(iterations), cost_num, 'r')
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Error vs Iterations')
    plt.show()

def accuracy(Ytrue,Ypred):
  acc=0;
  for i in range(len(Ytrue)):
    if Ytrue[i] == Ypred[i]:
      acc+=1;
  return acc/len(Ytrue)
def compute_predictions(X, W):
  lincom=W[0]+W[1]*X#+W[2]*X[:,1]#+W[3]*X.iloc[:,2]
  exp_component=np.exp(-lincom)
  #exp_component = [np.exp(-((W[1] * xi) + W[0])) for xi in X]    
  y_fit =  [e/(1 + e) for e in exp_component] 
  return y_fit
    
```
